```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

class Model_Evaluate ():
    
    def __init__(self, df):
        self.df = None
        self.df_train = None
        self.df_test = None
        self.knn = None
        self.test_prediction = None
    def set_dataset(self,df):
        self.df = df
    
    def KNN(self, train, test, n):
        
        # split data
        df_train, df_test = train_test_split(self.df,train_size =train, test_size = test, random_state=0)

        # Model KNN
        knn = KNeighborsClassifier(n_neighbors = n,metric = 'euclidean') # n_neighbors = k
        knn.fit(df_train[['latitude', 'longitude']],df_train['Abnormality'])
        
        #set attribute
        self.df_train = df_train
        self.df_test = df_test
        self.knn = knn
        self.test_prediction = knn.predict(self.df_test[['latitude', 'longitude']])
    
    def KnnEval(self):        

        accuracy = round(accuracy_score(self.df_test['Abnormality'], self.test_prediction),3)
        precision = round(precision_score(self.df_test['Abnormality'], self.test_prediction, average = 'macro'),3)
        recall = round(recall_score(self.df_test['Abnormality'], self.test_prediction, average = 'macro'),3)
        f1 = round(f1_score(self.df_test['Abnormality'], self.test_prediction, average = 'macro'),3)

        KNN_eval = ('Accuracy : '+ str(accuracy)+'\n'+
        'Precision : '+ str(precision)+'\n'+
        'Recall : '+ str(recall)+'\n'+
        'F1 Score: '+ str(f1)+'\n')
        
        return KNN_eval
    
    def report(self):
        report = ('total data test = '+str(self.df_test.flight_id.count())+'\n'+
            'total normal data test = '+str(self.df_test[self.df_test['Abnormality'] == 0].flight_id.count())+'\n'+
            'total abnormal data test = '+str(self.df_test[self.df_test['Abnormality'] == 1].flight_id.count())+'\n'+
            '\ntotal data train = '+str(self.df_train.flight_id.count())+'\n'+
            'total normal data train = '+str(self.df_train[self.df_train['Abnormality'] == 0].flight_id.count())+'\n'+
            'total abnormal data train = '+str(self.df_train[self.df_train['Abnormality'] == 1].flight_id.count())+'\n'+'\n')
        
        return report
        
        
    def confusion_matrix(self):
        conf_matrix = confusion_matrix(self.df_test['Abnormality'], self.test_prediction)

        fig, ax = plt.subplots(figsize=(3.21, 2.71))
        ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
        for i in range(conf_matrix.shape[0]):
            for j in range(conf_matrix.shape[1]):
                ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size=12)

        plt.xlabel('Predictions', fontsize=12)
        plt.ylabel('Actuals', fontsize=12)
        plt.title('Confusion Matrix', fontsize=15)
        plt.savefig('Model_KNN dan Confusion matrix/Confusion_matrix/CM_.png')
        plt.close()
        return 'Model_KNN dan Confusion matrix/Confusion_matrix/CM_.png'
        
    
```
