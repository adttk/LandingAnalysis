```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

df_train = pd.read_csv("Model_KNN dan Confusion matrix/KNN/data_Train.csv")
df_test = pd.read_csv("Model_KNN dan Confusion matrix/KNN/data Training 361.csv")
```

```python
feature_train = df_train[['latitude','longitude']]
target_train = df_train['Abnormality']
feature_test = df_test[['latitude','longitude']]
target_test = df_test['Abnormality']
```

```python
scaler = StandardScaler()
scaler.fit(feature_train)

feature_train = scaler.transform(feature_train)
feature_test = scaler.transform(feature_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=5,metric = 'euclidean')
classifier.fit(feature_train, target_train)
prediction_test = classifier.predict(feature_test)
```

```python
pred_result = df_test
pred_result['prediction'] = prediction_test
pred_result. to_csv("Model_KNN dan Confusion matrix/KNN/Data_pred.csv")
pred_result.head()
```

```python
print(classification_report(target_test,prediction_test))
```

```python
plt.figure(figsize=(16, 8))
mkr = {0: 'o', 1: 'x'}
for kind in mkr:
    d = df_train[df_train.Abnormality==kind]
    plt.scatter(d.latitude, d.longitude, 
                s = 100,
                alpha = 0.7,
                c = '#589651', 
                marker = mkr[kind])
    
for kind in mkr:
    d = pred_result[pred_result.Abnormality==kind]
    plt.scatter(d.latitude, d.longitude, 
                s = 100, 
                c = '#3d3d3d', 
                marker = mkr[kind])
plt.xlabel('Latitude',
           weight='bold',
           size=12)

plt.ylabel('Longitude',
           weight='bold',
           size=12)
plt.savefig("Model_KNN dan Confusion matrix/KNN/Data_pred vs data_train.png")
plt.show()
```

```python
TP = 0
TN = 0
FP = 0
FN = 0

for i in pred_result.index :
    
    if(pred_result['prediction'][i] == pred_result['Abnormality'][i]):
        if(pred_result['prediction'][i] == 1):
            TP += 1
        if(pred_result['prediction'][i] == 0):
            TN += 1
    else :
        if(pred_result['prediction'][i] == 1 and pred_result['Abnormality'][i] == 0):
            FP += 1
        if(pred_result['prediction'][i] == 0 and pred_result['Abnormality'][i] == 1):
            FN += 1
            
Accuracy = round((TP+TN)/(TP+FP+FN+TN),3)
Precision = round(TP/(TP + FP),3)
Recall  = round (TP /(TP + FN),3)
FScore  = round((2 * Recall * Precision) / (Recall + Precision),3)


print('Accuracy = ' + str(Accuracy))
print('Precision = ' + str(Precision))
print('Recall = '+ str(Recall))
print('FScore = '+str(FScore))
print('-------')
print(TP)
print(TN)
print(FP)
print(FN)
    
```
