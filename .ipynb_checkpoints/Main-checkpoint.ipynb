```python
# Visualisasi dataset
df = pd.DataFrame(a.data_with_abnormality)

plt.figure(figsize=(16, 8))
df_normal = df[df['Abnormality'] == '0']
plt.scatter(df_normal['latitude'], df_normal['longitude'], 
            s = 100,
            alpha = 0.7,
            c = '#3d3d3d')

df_abnormal = df[df['Abnormality'] == '1']
plt.scatter(df_abnormal['latitude'], df_abnormal['longitude'], 
            s = 100,
            alpha = 0.7,
            c = '#F22613')
    
plt.xlabel('Latitude',
           weight='bold',
           size=12)

plt.ylabel('Longitude',
           weight='bold',
           size=12)
plt.show()
```

```python
# split data
df_machine, df_unsign = train_test_split(df, test_size = 0.2, random_state=0)
df_unsign.to_csv('unsign.csv')
df_machine.to_csv('machine.csv')

df_train, df_test = train_test_split(df_machine, test_size = 0.4, random_state=0)

# Model KNN
knn = KNeighborsClassifier(n_neighbors = 3,metric = 'euclidean') # n_neighbors = k
knn.fit(df_train[['latitude', 'longitude']],df_train['Abnormality'])
test_prediction = knn.predict(df_test[['latitude', 'longitude']])

#Hasil prediksi
df_prediction = df_test.copy()
df_prediction['Predict'] = test_prediction
df_prediction.to_csv('prediction.csv')

# report split data
print('total data test = '+str(df_test.flight_id.count()))
print('total normal data test = '+str(df_test[df_test['Abnormality'] == '0'].flight_id.count()))
print('total normal data test = '+str(df_test[df_test['Abnormality'] == '1'].flight_id.count()))
print('\ntotal data train = '+str(df_train.flight_id.count()))
print('total normal data train = '+str(df_train[df_train['Abnormality'] == '0'].flight_id.count()))
print('total normal data train = '+str(df_train[df_train['Abnormality'] == '1'].flight_id.count()))

```

```python
conf_matrix = confusion_matrix(df_prediction['Abnormality'], df_prediction['Predict'])

fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.savefig('Model_KNN dan Confusion matrix/Confusion_matrix/CM_3.png')
plt.show()
```

```python
accuracy = round(accuracy_score(df_test['Abnormality'], test_prediction),3)
precision = round(precision_score(df_test['Abnormality'], test_prediction, average = 'macro'),3)
recall = round(recall_score(df_test['Abnormality'], test_prediction, average = 'macro'),3)
f1 = round(f1_score(df_test['Abnormality'], test_prediction, average = 'macro'),3)

print('Accuracy : '+ str(accuracy))
print('Precision : '+ str(precision))
print('Recall : '+ str(recall))
print('F1 Score: '+ str(f1))
```
