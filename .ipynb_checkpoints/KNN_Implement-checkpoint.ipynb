```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from Model_Evaluate import Model_Evaluate

class KNN_Implement():
    def __init__(self):
        self.df = None
        
    def set_data_train(self, df):
        self.df = df
        
    def FixModel(self):
        knn = KNeighborsClassifier(n_neighbors = 5,metric = 'euclidean') # n_neighbors = k
        knn.fit(self.df[['latitude', 'longitude']],self.df['Abnormality'])
        return knn
                                           
    def Predict(self, unseen):
        knn = self.FixModel()
        unseen_prediction = knn.predict(unseen[['latitude', 'longitude']])

        #Hasil prediksi
        df_prediction = unseen.copy()
        df_prediction['Abnormality'] = unseen_prediction
        df_prediction.to_csv('prediction_unseen.csv')
        return df_prediction
        
        
        
        
```

```python
df = pd.read_csv('clean_data.csv')
unseen = pd.read_csv('unseen_data.csv')
KNN = KNN_Implement()
KNN.set_data_train(df)
KNN.Predict(unseen)
```
