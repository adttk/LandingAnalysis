```python
import pandas as pd
import matplotlib.pyplot as plt

# Load table to DataFrame
df = pd.read_csv("Data_Preparation/Data_Grouping/step_2/Data_Abnormality.csv")

# Create pivot tabel about count of airlines
df = df.groupby('airline_icao')
df = df['airline_icao'].count()
df = pd.DataFrame(df)

# Save pivot DataFrame to csv file
df.to_csv("Data_Visualization/Airlines/Count_of_airlines_flight/Count_of_airline_flights.csv")

# Load table to DataFrame
df = pd.read_csv("Data_Visualization/Airlines/Count_of_airlines_flight/Count_of_airline_flights.csv")

# change attributes name
df=df.set_axis(['Airline icao' ,
                'Count of Flight'],
                axis ='columns')

# Save pivot DataFrame to csv file
df.to_csv("Data_Visualization/Airlines/Count_of_airlines_flight/Count_of_airline_flights.csv", index = False)

# Create data viz with bar chart
plt.bar(df['Airline icao'],
        df['Count of Flight'],
        width = 0.5)

plt.xlabel('Airlines Code',
           weight='bold',
           size=10)

plt.ylabel('Count',
           weight='bold',
           size=10)

# save viz to png file
plt.savefig('Data_Visualization/Airlines/Count_of_airlines_flight/Count_of_airline_flights.png')
plt.show()
```

```python
# Load table to DataFrame
df = pd.read_csv("Data_Preparation/Data_Grouping/step_2/Data_Abnormality.csv")

# Create pivot tabel about airlines and abnormality
df = df.groupby(['airline_icao','Abnormality'])['Abnormality'].aggregate('count').unstack()
df = pd.DataFrame(df)

# Save pivot DataFrame to csv file
df.to_csv("Data_Visualization/Airlines/Abnormality_by_all_flights/Abnormality_by_all_flights.csv")

# Load table to DataFrame
df = pd.read_csv("Data_Visualization/Airlines/Abnormality_by_all_flights/Abnormality_by_all_flights.csv")

# Change null data to 0
df = df.fillna(0)

# change attributes name
df=df.set_axis(['Airline icao' ,
                'Count of normal',
                'Count of abnormal',],
                axis ='columns')

# calculate normal and abnormal data to new attributes
total_normal= sum(df['Count of normal'])
total_abnormal= sum(df['Count of abnormal'])

# calculate the percentage normal and abnormal data for new attributes
df['Percentage normal of all normal data (%)'] = round((df['Count of normal']/total_normal)*100, 2)
df['Percentage abnormal of all abnormal data (%)'] = round((df['Count of abnormal']/total_abnormal)*100,2)

# Save DataFrame to new csv file
df.to_csv("Data_Visualization/Airlines/Abnormality_by_all_flights/Abnormality_by_all_flights.csv")

# Create data viz with table heatmap
df = df.style.background_gradient(
    cmap='Greens',
    subset= 'Percentage normal of all normal data (%)').background_gradient(
                             cmap='Reds',
                             subset= 'Percentage abnormal of all abnormal data (%)')

#Save table heatmap to xlsx file
df.to_excel("Data_Visualization/Airlines/Abnormality_by_all_flights/heatmap1.xlsx")
df
```

```python
df = pd.read_csv("Data_Preparation/Data_Grouping/step_2/Data_Abnormality.csv")

df = df.groupby(['airline_icao','Abnormality'])['Abnormality'].aggregate('count').unstack()
df = pd.DataFrame(df)

df.to_csv("Data_Visualization/Airlines/Abnormality_by_airline_flight/Abnormality_by_airline_flights.csv")
df = pd.read_csv("Data_Visualization/Airlines/Abnormality_by_airline_flight/Abnormality_by_airline_flights.csv")
df=df.fillna(0)
df=df.set_axis(['Airline icao' ,
                'Count of normal',
                'Count of abnormal'],
                axis ='columns')


df['Percentage normal of airline data (%)'] = round((df['Count of normal']/(df['Count of abnormal']+df['Count of normal']))*100,2)
df['Percentage abnormal of airline data (%)'] = round((df['Count of abnormal']/(df['Count of abnormal']+df['Count of normal']))*100,2)

df.to_csv("Data_Visualization/Airlines/Abnormality_by_airline_flight/Abnormality_by_airline_flights.csv")

df = df.style.background_gradient(
    cmap='Greens',
    subset= 'Percentage normal of airline data (%)').background_gradient(
                             cmap='Reds',
                             subset= 'Percentage abnormal of airline data (%)')

df.to_excel("Data_Visualization/Airlines/Abnormality_by_airline_flight/heatmap2.xlsx")
df
```
